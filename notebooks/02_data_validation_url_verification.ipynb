```python
from google.colab import drive
drive.mount('/content/drive')

base_dir = "/content/drive/MyDrive/Colab Notebooks/PFA"
data_dir = f"{base_dir}/data"
json_path = f"{base_dir}/incident_subset_1000.json"
```

```python
import json
import os
from PIL import Image
from tqdm import tqdm

# Charger le JSON
with open(json_path, 'r') as f:
    data = json.load(f)

# Récupérer uniquement les images valides existantes
valid_images = []
target_categories = {"on fire", "earthquake", "heavy rainfall", "fog"}

def is_valid_image(path):
    try:
        img = Image.open(path)
        img.verify()
        return True
    except:
        return False

for img_info in tqdm(data):
    category = img_info["category"]
    if category not in target_categories:
        continue

    filename = os.path.basename(img_info["filename"])
    local_path = os.path.join(data_dir, category.replace(" ", "_"), filename)

    if os.path.exists(local_path) and is_valid_image(local_path):
        valid_images.append((local_path, category))
    else:
        print(f"❌ Image manquante ou corrompue : {local_path}")

print(f"✅ {len(valid_images)} images valides trouvées")
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd

# Séparer les chemins et labels
filepaths, labels = zip(*valid_images)

# Split 70/20/10
train_files, temp_files, train_labels, temp_labels = train_test_split(
    filepaths, labels, test_size=0.3, random_state=42, stratify=labels
)
val_files, test_files, val_labels, test_labels = train_test_split(
    temp_files, temp_labels, test_size=1/3, random_state=42, stratify=temp_labels
)

# Sauvegarder en CSV avec suffixe '_f'
def save_csv(name, paths, labels):
    df = pd.DataFrame({'filepath': paths, 'label': labels})
    df.to_csv(f"{base_dir}/{name}_f.csv", index=False)

save_csv("train", train_files, train_labels)
save_csv("val", val_files, val_labels)
save_csv("test", test_files, test_labels)

print("✅ Nouveaux fichiers filtrés : train_f.csv, val_f.csv, test_f.csv")
```
